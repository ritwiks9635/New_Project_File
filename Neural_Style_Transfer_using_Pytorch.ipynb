#**Loading VGG Pretrained Model**

```python
import torch
from torchvision import models

vgg = models.vgg19(pretrained = True)
print(vgg)
```

```python
# in neural style transfer, we dont need classifier part since we need only content and style
vgg = vgg.features
print(vgg)
```

```python
# dont need to update weights, no gradient computation
for parameters in vgg.parameters():
  parameters.requires_grad_(False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
vgg.to(device)
```


#**Preprocess image**
Torchvision models page : https://pytorch.org/docs/stable/torchvision/models.html

```python
from PIL import Image
from torchvision import transforms as T

def preprocess(img_path, max_size=500):
  image = Image.open(img_path).convert('RGB')

  if max(image.size) > max_size:
    size = max_size
  else:
    size = max(image.size)

  img_transforms = T.Compose([
              T.Resize(size),
              T.ToTensor(),
              T.Normalize(mean=[0.485, 0.456, 0.406],
                          std=[0.229, 0.224, 0.225]) # https://pytorch.org/vision/0.11/models.html
  ])

  image = img_transforms(image)

  # add batch size: (c, h, w) -> (m, c, h, w)
  # img is already (c, h, w) since use torchvision.transforms
  image = image.unsqueeze(0)

  return image
```

```python
content_p = preprocess("/content/Project-NST/spiderman.jpg")
style_p = preprocess("/content/Project-NST/style10.jpg")

content_p = content_p.to(device)
style_p = style_p.to(device)

print("content shape is :: ", content_p.shape)
print('style shape is :: ', style_p.shape)
```

#**Deprocess image**

```python
import numpy as np
import matplotlib.pyplot as plt

# Do reverse from preprocess
# to plot img
def deprocess(tensor):
  image = tensor.to('cpu').clone()
  image = image.numpy()
  image = image.squeeze(0)  # (m, c, h, w) -> (c, h, w)
  image = image.transpose(1, 2, 0)  #(c, h, w) -> (h, w, c)
  image = image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
  image = image.clip(0, 1)

  return image
```

```python
content_d = deprocess(content_p)
style_d = deprocess(style_p)

print("Deprocess contact shape ::", content_d.shape)
print("Deprocess style shape ::", style_d.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 10))

ax1.imshow(content_d)
ax2.imshow(style_d)
```

#**Get content,style features and create gram matrix**

```python
def get_features(image, model):
  layers = {
      '0': 'conv1_1',
      '5': 'conv2_1',
      '10': 'conv3_1',
      '19': 'conv4_1',
      '21': 'conv4_2',  # content_feature
      '28': 'conv5_1'
  }

  x = image
  Features = {}

  for name, layer in model._modules.items():
    x = layer(x)

    if name in layers:
      Features[layers[name]] = x

  return Features
```

```python
content_f = get_features(content_p, vgg)
style_f = get_features(style_p, vgg)
```

```python
def gram_matrix(tensor):
  b, c, h, w = tensor.size()
  tensor = tensor.view(c, h*w)  # unroll
  gram = torch.mm(tensor, tensor.t()) # matrix multiplication w/ its transpose
  return gram
```

```python
style_grams = {layer: gram_matrix(style_f[layer]) for layer in style_f}
```

#**Creating Style and Content loss function**

```python
def content_loss(target_conv4_2, content_conv4_2):
  loss = torch.mean((target_conv4_2 - content_conv4_2)**2)
  return loss
```

```python
style_weights = {
    'conv1_1': 1.0,
    'conv2_1': 0.75,
    'conv3_1': 0.2,
    'conv4_1': 0.2,
    'conv5_1': 0.2
}
```

```python
def style_loss(style_weights, target_features, style_grams):
  loss = 0.0

  for layer in style_weights:
    target_f = target_features[layer]
    target_gram = gram_matrix(target_f)

    style_gram = style_grams[layer]

    layer_loss = style_weights[layer] * torch.mean((target_gram - style_gram)**2)

    b, c, h, w = target_f.shape
    loss += layer_loss/(c*h*w)

  return loss
```

```python
target = content_p.clone().requires_grad_(True).to(device)
target_f = get_features(target, vgg)
print('Content loss:', content_loss(target_f['conv4_2'], content_f['conv4_2']))
print('Style loss:', style_loss(style_weights, target_f, style_grams))
```

#**Training loop**

```python
from torch import optim

optimizer = optim.Adam([target], lr=0.003)

alpha = 1
beta = 1e5

epochs = 3000
show_every = 500
```

```python
def total_loss(c_loss, s_loss, alpha, beta):
  loss = alpha*c_loss + beta*s_loss
  return loss
```

```python
results = []

for i in range(epochs):
  target_f = get_features(target, vgg)

  c_loss = content_loss(target_f['conv4_2'], content_f['conv4_2'])
  s_loss = style_loss(style_weights, target_f, style_grams)
  t_loss = total_loss(c_loss, s_loss, alpha, beta)

  optimizer.zero_grad()
  t_loss.backward()
  optimizer.step()

  if i % show_every == 0:
    print('Total loss at epoch %d: %f'%(i, t_loss))
    results.append(deprocess(target.detach()))
```

```python
plt.figure(figsize=(10, 8))

for i in range(len(results)):
  plt.subplot(3, 2, i+1) # 3row-2cols
  plt.imshow(results[i])

plt.show()
```

```python
target_copy = deprocess(target.detach())
content_copy = deprocess(content_p)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(target_copy)
ax2.imshow(content_copy)
```
